```python
#using LinearRegression API

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

dataset = pd.read_csv(r'canada_pci.csv')
X = dataset.iloc[:, :-1].values 
y = dataset.iloc[:, 1].values 

from sklearn.model_selection import train_test_split 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=1/3, random_state=0)

from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)

y_pred = regressor.predict(X_test)
print(y_pred)

viz_train = plt
viz_train.scatter(X_train, y_train, color='red')
viz_train.plot(X_train, regressor.predict(X_train), color='green')
viz_train.title('Salary VS Experience ')
viz_train.xlabel('Year of Experience')
viz_train.ylabel('Salary')
viz_train.show()
```

```python
#using basic math

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

def estimate_coeff(x, y):
    n = np.size(x)
    m_x = np.mean(x)
    m_y = np.mean(y)

    SS_xy = np.sum(y * x) - n * m_y * m_x
    SS_xx = np.sum(x * x) - n * m_x * m_x

    b_1 = SS_xy / SS_xx
    b_0 = m_y - b_1 * m_x

    return (b_0, b_1)

def plot_regression_line(x, y, b):
    plt.scatter(x, y, color="m", marker="o", s=30)

    y_pred = b[0] + b[1] * x

    plt.plot(x, y_pred, color="g")

    plt.xlabel('x')
    plt.ylabel('y')

    plt.show()

def main():
    # Step 1: Load the dataset from a CSV file
    data = pd.read_csv('salary_data.csv')

    # Step 2: Prepare the data
    x = data['YearsExperience']
    y = data['Salary']

    # Step 3: Estimate coefficients
    b = estimate_coeff(x, y)
    print("Estimated coefficients:\nb_0 = {}\nb_1 = {}".format(b[0], b[1]))

    # Step 4: Plot regression line
    plot_regression_line(x, y, b)


if __name__ == "__main__":
    main()

```
